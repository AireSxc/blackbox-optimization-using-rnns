```python
%matplotlib inline
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import axes3d
```

```python
def kamel(x):
    return (4-2.1*x[0,:]**2+x[0,:]**4/3)*x[0,:]**2+x[0,:]*x[1,:]+(-4+4*x[1,:]**2)*x[1,:]**2

def ackley(x):
    n = x.shape[0]
    return 20+np.exp(1)-20*np.exp(-0.2*np.sqrt(1/n*np.sum(D**2,0)))-np.exp(1/n*np.sum(np.cos(2*np.pi*D),0))

def levi(x):
    y = 1+(x-1)/4
    return np.sin(np.pi*y[0,:])**2+np.sum((y[0:-1]-1)**2*(1+10*np.sin(np.pi*y[0:-1]+1)**2),0)+(y[-1,:]-1)**2*(1+np.sin(2*np.pi*x[-1,:])**2)

def quad(x):
    n = x.shape[0]
    s = 0
    for i in range(n):
        s += np.sum(x[0:i,:],0)**2
    return s

def rastigrin(x):
    n = x.shape[0]
    return 10*n+np.sum(x**2,0)/4000-np.prod(np.cos(x/np.sqrt(np.linspace(1,n,n)).reshape(n,1,1)),0)+1

def rosenbrock(x):
    xp1 = x[1:,:]
    x = x[0:-1,:]
    return np.sum(100*(x**2-xp1)**2+(x-1)**2,0)

def schwefel(x):
    return np.sum(np.abs(x),0)+np.prod(np.abs(x),0)

def sphere(x):
    return np.sum(x**2,0)

def sumsquares(x):
    n = x.shape[0]
    return np.sum(np.linspace(1,n,n).reshape(n,1,1)*x**2,0)
```

```python
xx = np.linspace(-10,10,201)
yy = np.linspace(-10,10,201)

X, Y = np.meshgrid(xx,yy)
D = np.array((X,Y))
```

```python
fig = plt.figure(figsize=(20,20))

ax1 = fig.add_subplot(331, projection='3d')
surf = ax1.plot_surface(X, Y, ackley(D), cmap=cm.coolwarm, 
                       linewidth=0, antialiased=False)
plt.title("Ackley")


ax2 = fig.add_subplot(332, projection='3d')
surf = ax2.plot_surface(X, Y, levi(D), cmap=cm.coolwarm, 
                       linewidth=0, antialiased=False)
plt.title("Levi")

ax2 = fig.add_subplot(333, projection='3d')
surf = ax2.plot_surface(X, Y, quad(D), cmap=cm.coolwarm, 
                       linewidth=0, antialiased=False)
plt.title("Quadric")

ax2 = fig.add_subplot(334, projection='3d')
surf = ax2.plot_surface(X, Y, rastigrin(D), cmap=cm.coolwarm, 
                       linewidth=0, antialiased=False)
plt.title("Rastigrin")

ax2 = fig.add_subplot(335, projection='3d')
surf = ax2.plot_surface(X, Y, rosenbrock(D), cmap=cm.coolwarm, 
                       linewidth=0, antialiased=False)
plt.title("Rosenbrock")

ax2 = fig.add_subplot(336, projection='3d')
surf = ax2.plot_surface(X, Y, schwefel(D), cmap=cm.coolwarm, 
                       linewidth=0, antialiased=False)
plt.title("Schwefel")

ax2 = fig.add_subplot(337, projection='3d')
surf = ax2.plot_surface(X, Y, sphere(D), cmap=cm.coolwarm, 
                       linewidth=0, antialiased=False)
plt.title("Sphere")

ax2 = fig.add_subplot(338, projection='3d')
surf = ax2.plot_surface(X, Y, sumsquares(D), cmap=cm.coolwarm, 
                       linewidth=0, antialiased=False)
plt.title("Sum Squares")

ax2 = fig.add_subplot(339, projection='3d')
surf = ax2.plot_surface(X, Y, kamel(D), cmap=cm.coolwarm, 
                       linewidth=0, antialiased=False)
plt.title("Kamel")



plt.show()
```
